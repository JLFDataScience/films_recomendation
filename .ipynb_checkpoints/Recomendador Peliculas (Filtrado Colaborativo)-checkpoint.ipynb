```python
import pandas as pd
import numpy as np

#Cargamos el fichero de películas con sus genéros. movies.dat tiene los generos de las películas separados por barra
peliculas = pd.read_table('data/movies.dat', header=None, sep='::', engine='python', names=['movie_id', 'movie_title', 'movie_genre'])

peliculas.head()
```

```python
ratings = pd.read_table('data/ratings.dat', header=None, sep='::', engine='python', names=['user_id', 'movie_id', 'rating', 'timestamp'])
ratings.head()
```

```python
#Cargamos el fichero con las puntuaciones de las películas, ratings.dat
ratings = pd.read_table('data/ratings.dat', header=None, sep='::', engine='python', names=['user_id', 'movie_id', 'rating', 'timestamp'])

#Borramos la fecha en la que el rating fue creado
del ratings ['timestamp']

#Reemplazamos la id de la película por su título para tener una mayor claridad
ratings = pd.merge(ratings, peliculas, on='movie_id')[['user_id', 'movie_title', 'movie_id','rating']]

ratings.head()
```

```python
# Creamos una nueva matriz con los ratios de cada usuario para todas las películas
ratings_matriz = ratings.pivot_table(values='rating', index='user_id', columns='movie_title')
# Ponemos 0 donde no tengamos puntuación
ratings_matriz.fillna(0, inplace=True)

pelicula_index = ratings_matriz.columns

ratings_matriz.head()
```

```python
pelicula_index
```

```python
# Calculamos la correlación (coeficiente de correlación de Pearson(PMCC)) entre las películas en función de la puntuación de los usuarios.
# PMCC tiene un valor entre -1 y 1 que mide cuán relacionadas están un par de variables cuantitativas.
corr_matrix = np.corrcoef(ratings_matriz.T)
corr_matrix.shape
```

```python
favorita_pelicula = 'Toy Story (1995)'

favorita_pelicula_index = list(pelicula_index).index(favorita_pelicula)

P = corr_matrix[favorita_pelicula_index]

#Mostramos las películas más relacionadas con Toy Story según su valor de correlación.
list(pelicula_index[(P>0.4) & (P<1.0)])
```

```python
#determina la posición en el listado de películas que tiene  Toy Story
list(pelicula_index).index(favorita_pelicula)
```

```python
#El listado de correlación que tiene la película Toy Story con cada uno de los usuarios
P
```

```python
#Devuelve el vector de correlación para una película
def get_pelicula_similar(pelicula):
    movie_idx = list(pelicula_index).index(pelicula)
    return corr_matrix[movie_idx]

#Devolvemos las películas más similares a los gustos de un usuario determinado.
#Si queremos recomendar películas a un usuario, obtenemos la lista de las películas que ha visto y sumamos las correlaciones
#de esas películas con todas las demás para devolver las películas con una mayor correlación total.
def get_pelicula_recomendaciones(usuario):
    pelicula_similares = np.zeros(corr_matrix.shape[0])
    for pelicula_id in usuario:
        pelicula_similares = pelicula_similares + get_pelicula_similar(pelicula_id)
        similares_df = pd.DataFrame({
        'Titulo': pelicula_index,
        'sum_similar': pelicula_similares
        })
    similares_df = similares_df[~(similares_df.Titulo.isin(usuario))]
    similares_df = similares_df.sort_values(by=['sum_similar'], ascending=False)
    return similares_df
```

```python
#Vamos a recomendar películas al usuario 21, que ha visto las siguientes películas con su rating correspondiente
sample_usuario = 21
ratings[ratings.user_id==sample_usuario].sort_values(by=['rating'], ascending=False)
```

```python
#Determinamos el listado de títulos de películas dei usuario 21. Se coge el df ratings y se convierte a listado la columna
#movie_title
sample_usuario_peliculas = ratings[ratings.user_id==sample_usuario].movie_title.tolist()
sample_usuario_peliculas
```

```python
sample_usuario_peliculas = ratings[ratings.user_id==sample_usuario].movie_title.tolist()
#Mandamos a la funcion get_pelicula_recomendaciones el listado de películas que ha visto el usuario para que nos de el listado
#de películas recomendadas.
recomendaciones = get_pelicula_recomendaciones(sample_usuario_peliculas)

#Obtenemos las 20 películas con mejor puntuación
recomendaciones.Titulo.head(20)
```
